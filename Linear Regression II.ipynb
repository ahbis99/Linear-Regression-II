```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

import warnings
warnings.filterwarnings("ignore")
```

```python
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

1. Hedef değişkeninizin görünür sıcaklık ve sıcaklık arasındaki fark olduğu doğrusal bir regresyon modeli oluşturun. Açıklayıcı . değişkenler olarak nem ve rüzgar hızı kullanın. Şimdi, modelinizi OLS kullanarak tahmin edin. Tahmin edilen katsayılar istatistiksel olarak anlamlı mıdır? Tahmini katsayılar önceki beklentileriniz doğrultusunda mı? Tahmin edilen katsayıları yorumlayınız. Hedef ve açıklayıcı değişkenler arasındaki ilişkiler nelerdir?

```python
X = weather_df[["Humidity","Wind Speed (km/h)"]]
Y = weather_df["Temperature (C)"] - weather_df["Apparent Temperature (C)"]

lrm = LinearRegression()
lrm.fit(X, Y)

print("Temperature (C)")
x = sm.add_constant(X)
results = sm.OLS(Y, x).fit()
results.summary()
```

- When we look at the P values, we see that it is close to 0 in both models, so we can say that the coefficients are statistically significant.
- If we look at the coefficients, it has a increasing effect on the temperature in our two variables. We see that the temperature effect of humidity is more, which is also true as expected.

2.  Ardından, yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin. Katsayılar istatistiksel olarak anlamlı mıdır? Nem ve rüzgar hızı için tahmini katsayıların işaretleri değişti mi? Tahmin edilen katsayıları yorumlayınız.

```python
weather_df["Hum_wind"] = weather_df["Humidity"] * weather_df["Wind Speed (km/h)"]
new_X = X = weather_df[["Humidity","Wind Speed (km/h)","Hum_wind"]]

lrm = LinearRegression()
lrm.fit(new_X, Y)

x = sm.add_constant(new_X)
results = sm.OLS(Y, x).fit()
results.summary()
```

- The estimated coefficients are statistically significant. The coefficients for wind speed and humidity decreased and their signs changed. Interacting variable has more effect on target variable now.

```python
df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
df.head()
```

1. Ev fiyatları modelinizi tekrar çalıştırın ve sonuçları yorumlayın. Hangi özellikler istatistiksel olarak anlamlı ve hangileri değildir?

```python
dummies = pd.get_dummies(df[["SaleType","Street"]],drop_first=True)
X = pd.concat([df["FullBath"],dummies],axis=1,ignore_index=False)
Y = df["SalePrice"]

lrm = LinearRegression()
lrm.fit(X, Y)

x = sm.add_constant(X)
results = sm.OLS(Y, x).fit()
results.summary()
```

- When we examine the OLS, we see that many dummy variables are not statistically significant because the p values are not close to zero.

```python
X = X[["FullBath","SaleType_Con","SaleType_New","SaleType_WD"]]

lrm = LinearRegression()
lrm.fit(X , Y)

x = sm.add_constant(X)
results = sm.OLS(Y, x).fit()
results.summary()
```

- We can say that the constant changes noticeably and there is also a change in the coefficients of the features.
